```python
#Define a path to some data to test out functions
testDataPath = '../testData/rasterData/OD_10m_MicroTest.tif'

#Load the grid in, grids 'require' 
testDEM = dpy.demGrid([],rasterPath = testDataPath)

#Check to make sure it looks right
testDEM.plotGrid()
```

```python
'''
We can directly mathematical operations on demGrid objects, here we multiply by progressively larger values
and plot the results.
'''

#First create some axes for storing plots.
f,axs = plt.subplots(1,2,figsize = (8,6),dpi = 120)

#Actual values are stored in '.grid' which can be accessed for desired operations
minZ = testDEM.min() #np.min(testDEM.grid)
maxZ = testDEM.max() #np.max(testDEM.grid)

#Plotting functions are able to accept any keyword arguments that would normally be based to pyplot.imshow
testDEM.plotGrid(axs = axs[0],vmin = minZ, vmax = maxZ)

#Mathematical operations can be conducted as normal on a grid - but warning, there is no checking for spatial overlap
#These operations will produce a new demGrid instance
newGrid = testDEM * 1.15
newGrid.plotGrid(axs = axs[1],vmin = minZ, vmax = maxZ)

# #Mathematical operations can also be used to modify grid values in place
# testDEM*=1.2
# testDEM.plotGrid(axs = axs[2],vmin = minZ,vmax = maxZ)

plt.tight_layout()
```

```python
#Many standard topographic analyses can be called on a demGrid
f,axs = plt.subplots(1,2,figsize = (6,4), dpi = 120)

#Calculate the magnitude of slope
slopeMag = testDEM.calcSlopeMagnitude()
slopeMag.plotGrid(axs = axs[0],vmin = 0,vmax = 0.5)

#Calculate the laplacian of the surface
L = testDEM.calcFiniteLaplacian()
L.plotGrid(axs = axs[1], vmin = -0.025, vmax = 0.025)

plt.tight_layout()
```

```python
#Plotting on the same axis allows for overlays, by using the 'alpha' argument to adjust transparency
f,axs = plt.subplots(1,1, figsize = (6,6), dpi = 300)

#Can calculate finite differences over broader kernels to look at longer wavelength slopes
slopeMag = testDEM.calcFiniteSlopeMagOverWindow(N = 3)

hs = testDEM.calcHillshade(azimuth = 45, altitude=45)

slopeMag.plotGrid(axs = axs,vmin = 0.0, vmax = 0.6,cmap = 'Greys')
hs.plotGrid(axs = axs,vmin = 0, vmax = 255, cmap = 'Greys_r')#, alpha = 0.4)

axs.set_xticklabels([])
axs.set_yticklabels([])

plt.savefig('/Users/sjohnstone/OneDrive - DOI/Documents/Publications/In Prep/DrainageOrientations_OzarkDome/Figures/ExplorationFigures/SpectralAnomaliesDescription/SlopeShade_repeatingDrainainges.png',
           transparent = True)
```

```python
'''
We can also create masked versions of grids (all values outside the mask are set to np.nan). This can
be a way to look at data as a function of (for example) aspect. 
'''

#First lets calculate aspect
aspectGrid = testDEM.calcAzimuth(N=3, returnAsArray = False) #If an 'N' is provided, finite differences will be calculated over a window

#Lets mask out the N facing and south facing data
angleHalfWindow = 45.0 #How many degrees about true north/ south to look?

isNFacing = (aspectGrid > (360 - angleHalfWindow)) | (aspectGrid < angleHalfWindow)
isSFacing = (aspectGrid < (180 + angleHalfWindow)) & (aspectGrid > (180 - angleHalfWindow))

#Now lets get out slopes
slope = testDEM.calcFiniteSlopeMagOverWindow(N = 3)

#Get masked grids for the north and south facing slopes
nfSlopes = slope.returnMaskedGrid(isNFacing)
sfSlopes = slope.returnMaskedGrid(isSFacing)

#Lets plot histograms of these data
f,axs = plt.subplots(1,1)

#The plot histogram function will take an axis to plot on and any arguments to pass through to pyplot.histogram
nfSlopes.plotHistogram(axs = axs, histtype = 'stepfilled',density = True,bins = 50,
                       color = 'dodgerblue', label = 'North facing')
sfSlopes.plotHistogram(axs = axs, histtype = 'stepfilled', density = True,bins = 50,
                       color = 'mediumseagreen', alpha = 0.65, label = 'South facing')

axs.set_xlabel('Slope')
axs.legend()
```

```python
#We can save grids as raster's (uses gdal on the backend).
savePath = '../testData/rasterData/test_hillshadeSave.tif'

'''
The save function takes a file path, and some information to tell GDAL what type of file to save as (GDALDRIVERNAME), 
and what datatype to save data as GDALDATATYPE. The default if these aren't specified are geotiffs, and float32. 
To see options check: 

https://gdal.org/drivers/raster/index.html - looking for the 'shortname' column for GDALDRIVERNAME

https://gdal.org/doxygen/gdal_8h.html#a22e22ce0a55036a96f652765793fb7a4 - lists optional datatypes,

reference these here by their (0 indexed) positions in this list
'''

hs.saveGrid(savePath,GDALDRIVERNAME = 'GTiff', GDALDATATYPE=6)
```

```python
'''Objects loaded from grids will remember the path of the file that created them. When some of the
pre-programmed derivatives are calculated, the file path for this derivatives will be updated, so that its easy to
resave this grid. This is done with the .overwriteSaveFile method '''
newHS = testDEM.calcHillshade()

print('This grid will default save to: {}'.format(newHS._filePath))
newHS.overwriteSaveFile()
```

```python
'''To update the file name by appending a new suffix, use the .updateFileNameWithSuffix function'''
newHS.updateFileNameWithSuffix('_newSufx')

print('This grid will default save to: {}'.format(newHS._filePath))
```

```python
'''In addition to map style plots, we can plot grids against one another. Here we will plot curvature as a function
of slope. No real expectations from this plot, but it shows the functionality available...'''

#First lets calculate these grids, over a decent window so that they are somewhat smooth
slopeGrid = testDEM.calcFiniteSlopeMagOverWindow(N = 4)
crvGrid = testDEM.calcFiniteLaplacianOverWindow(N = 4)

#Lets set some parameters for these pots
fractionOfDataToPlot = 0.25 #Let only plot ~25% of data, to simplify the plots a bit

#We can specify some axis labels, otherwise we'll be 'reminded' to
xAxisLabel = r'$\frac{dz}{dx}$' #This will take latex input because matplotlib
yAxisLabel = r'$\nabla ^2 z$'

#Create a plot axis
f,axs = plt.subplots(1,1)

#Plot all the grid observations there
axs = crvGrid.plotThisGridAgainstAnotherGrid(slopeGrid, plotFractionOfObservations = fractionOfDataToPlot,
                                       axs = axs, xAxisLabel = xAxisLabel,yAxisLabel = yAxisLabel, 
                                      color = 'k', markersize = 4, alpha = 0.1, zorder = -1) #Optional plot args

#We can also plot binned mean observations in a similar way
axs = crvGrid.plotMedianValuesBinnedByOtherGrid(slopeGrid,bins = 20,axs = axs,
                                                color = 'orange',zorder = 2) #zorder controls staking of plotted items

#Data are pretty scattered - lets look in log space
# axs.set_xscale('log')
# axs.set_yscale('symlog') #Symlog b/c curvatures can be negative

#Sometimes axis fonts are a littled small - can always update these from the axis
axs.xaxis.label.set_fontsize(14)
axs.yaxis.label.set_fontsize(14)

plt.tight_layout()
```
