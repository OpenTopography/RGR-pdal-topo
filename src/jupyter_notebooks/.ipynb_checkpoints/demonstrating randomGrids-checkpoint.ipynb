```python
import sys
sys.path.append('../demAnalysisComponents/') #Add the path to the demAnalysis scripts

import dem as dpy
import flowRoutingGrids as flw
import randomGrids as rg
from matplotlib import pyplot as plt
import numpy as np

from importlib import reload #While testing, this can be helpful to have loaded in

%matplotlib notebook
```

```python
#First, lets open an example real DEM just to remind ourselves what we are working with
#Define a path to some data to test out functions
testDataPath = '../testData/rasterData/OD_10m_microTest.tif'

#Load the grid in
testDEM = dpy.demGrid([],rasterPath = testDataPath)

#Check to make sure it looks right
testDEM.plotGrid()
```

```python
#In looking for significant features in landscapes, it can be tricky to define a null hypothesis.
#The randomGrids module provides some routines for generating topography
import randomGrids as rg

reload(rg)

#Here we will use the 'diamond square' algorithm to produce a dem of the same size as this example topography

#This algorithm is designed to produce random fractal topography based on a roughness value
#and some overall topographic variance (which we will set based on the actual topography)

#Decide on a roughness value between 0 and 1 (larger values = smoother topography)
roughnessValue = 0.7

#The random topography can be made consistent through specification of a random seed
randSeed = 42

#We can force the resulting grid to match the mean elevation and relief of this landscape, if we don't
#its relief will be similar, but its mean elevation will be close to zero
forceMatchMeanElevRelief = False

#Create an instance of this procedural grid generation routine, storing out preferences for grid creation
dsq = rg.proceduralGrid_diamondSquare(testDEM,roughness = roughnessValue,randomSeed= 20,
                                      matchMeanElevationAndRelief=forceMatchMeanElevRelief)

#Use that instance to create a single randomly generated topography
randInst = dsq.getPermutation()

#Set up an axis, and plot this randomly generated topography on that axis
f,ax = plt.subplots(1,1)
randInst.plotGrid(axs = ax)

# Get the images on an axis
im = ax.images[-1]        

#Get the colorbar for that image
plt.colorbar(im)
plt.tight_layout()
```

```python
#Lets explore how the roughness value impacts the resulting topography by looping through a few roughness values

#How many roughness values do we want to test
nValsToTest = 5

#Create a list of roughness values that we would like to look at
roughnessToTest = np.linspace(0,1,nValsToTest)

#Create an axis to plot the results on
f,axs = plt.subplots(1,nValsToTest,sharex=True, sharey=True)

#Lets make sure all the plots have the same color axis range
minZ = testDEM.min()
maxZ = testDEM.max()


#Loop through the roughness values we wanted to test
for i,H_i in enumerate(roughnessToTest):
    
    #Create an instance of the random grid generation routine with this roughness
    dsq = rg.proceduralGrid_diamondSquare(testDEM,roughness = H_i,matchMeanElevationAndRelief=True)
    
    #Get a single random grid based on the specified parameters
    randInst = dsq.getPermutation()
    
    randInst.plotGrid(axs = axs[i],vmin = minZ,vmax = maxZ)
    
    #Remove the axis labels to reduce cluttering
    axs[i].set_xticklabels([])
    axs[i].set_yticklabels([])
    
    #Set the title so we know what we're looking at
    axs[i].set_title('H= {:.1f}'.format(H_i),fontsize = 'x-small')
```

```python
#To get a sense of how randomness impacts this, lets make a grid with one roughness value, but a handfull of realizations

gridWidth = 150 #Lets make a small, non georeferenced grid simplicity

#Create a 'dummy' grid object to use as a reference
miniTest = dpy.demGrid(np.random.rand(gridWidth,gridWidth))

#The scale sets the standard deviation of the largest wavelength elements in the random diamond square grid
scale = 50
roughVal = 0.75
dsq = rg.proceduralGrid_diamondSquare(miniTest,roughness = roughVal, randomSeed=None,
                                      matchMeanElevationAndRelief=False, startingNoiseScale=scale)

#Lets make a matrix of some number of rows and colums or random grids
nTestsRows = 3
nTestsCols = 3
f,axs = plt.subplots(nTestsRows, nTestsCols, sharex = True, sharey = True)

#Iterate through these values
for row in range(nTestsRows):
    for col in range(nTestsCols):
        randInst = dsq.getPermutation()
        
        #Lets force all these to be zero mean before plotting
        randInst-=randInst.mean()
        hs = randInst.calcHillshade()

        randInst.plotGrid(axs = axs[row,col], vmin = -scale, vmax = scale)
        hs.plotGrid(axs = axs[row,col], vmin = 0, vmax = 255, cmap = 'Greys', alpha = 0.25)
        
        #Remove the axis labels to reduce cluttering
        axs[row,col].set_xticklabels([])
        axs[row,col].set_yticklabels([])


plt.tight_layout()
plt.savefig('Test.eps')
```

```python
'''There are also routines for generating procedural terrain for looking at patterns in flow routing grids.

These can help with the construction of null-hypothesis expectations for drainage network patterns. 
This is a procedural terrain generation routine that modifies the priority flood algorithm described in:

    Barnes, R., Lehman, C., and Mulla, D., 2014, Priority-flood: An optimal depression-filling and
     watershed-labeling algorithm for digital elevation models: Computers and Geosciences, v. 62, p. 117127,
      doi:10.1016/j.cageo.2013.04.024.

Specifically, the algorithm fills terrain upstream from outlets, with channels allowed to 'grow' upstream one
cell at a time based on randomly assigned priorities.  This approach was applied and described in:

    Johnstone, S.A., Finnegan, N.J., and Hilley, G.E., 2017, Weak bedrock allows north-south elongation of
     channels in semi-arid landscapes: Earth and Planetary Science Letters, v. 478, p. 150158,
      doi:10.1016/j.epsl.2017.08.037.


with an addition made in that work to iteratively correct large elevation discrepencies on either side of a
divide produced by flow paths with different integrated drainage area paths.

Here we will create an object to preform the random filling that can be used to create these
random drainage networks. There are two options for controlling how random priorities are assigned during
filling, controlled by the optional boolean parameter 'doScalePriorityByElevation'. 

If this is False (the default), priorities are entirely random, which allows for more variability
in drainage networks - but some of this variability is unnatural, resulting in large elevation discontinuities
at internal divides as long channels wrap back around on themselves. 

If this is True, priorities of growing a channel are scaled by the current elevation of that channel.
This reduces the chance of growth for long, high elevation portions of networks untill shorter
networks have a chance to 'catch up'. 

To highlight these differences lets show some examples below

'''
reload(rg)
reload(flw)

testDataPath = '../testData/rasterData/OD_10m_microTest.tif'
# testDataPath = '../testData/rasterData/OD_10mTest.tif'

#Load in a grid that is goind to serve as a template
testDEM = dpy.demGrid([],rasterPath = testDataPath)

#Preform the flow routing on that grid
testFlowGrd = flw.flowRoutingGrids(testDEM,forceReCalculation=False)

#First create an instance of the randomFilling class that assigns priorities to upstream filling randomly
randBasinMaker = rg.proceduralGrid_randomPriorityFilling(testFlowGrd,fillSlope = 1e-3,randomSeed=42,
                                                        matchMeanElevationAndRelief=False,
                                                         doScalePriorityByElevation=False)
#Second create an instance of the randomFilling class that assigns priorities to upstream filling randomly,
#but scaled by the current elevation
randBasinMaker_elevationScaled = rg.proceduralGrid_randomPriorityFilling(testFlowGrd,fillSlope = 1e-3,randomSeed=42,
                                                        matchMeanElevationAndRelief=False,
                                                         doScalePriorityByElevation=True)
#Get a random flow grid from each of these
rndFlowGrid = randBasinMaker.getFlowGridPermutation()
rndFlowGrid_elevationScaled = randBasinMaker_elevationScaled.getFlowGridPermutation()

f,axs = plt.subplots(2,2,figsize = (8,8),sharex = True, sharey = True,dpi = 120)

axs[0,0].set_title('Random priority, elevation')
rndFlowGrid.plotGrid(axs = axs[0,0])

axs[1,0].set_title('Random priority, drainage area')
rndFlowGrid.plotDrainageArea(axs = axs[1,0])

axs[0,1].set_title('Random priority (elevation scaled)')
rndFlowGrid_elevationScaled.plotGrid(axs = axs[0,1])
axs[1,1].set_title('Random priority (elevation scaled)')
rndFlowGrid_elevationScaled.plotDrainageArea(axs = axs[1,1])
```

```python
'''When looking at individual catchments, the drainage divides and catchment shape provide added constraints
on the geometry of resulting networks. Using the drainage basin as a mask imposes this contraint, though we also
need to specify where the outlet is located.

'''
reload(rg)
%matplotlib notebook

#Find the largest basin
outletIdx = np.argmax(testFlowGrd.areaGrid)
outletRowCol = np.unravel_index(outletIdx,testFlowGrd.shape) #Transform that index to row,col coordinates

# outlets = testFlowGrd.findBoundaryOutlets()
basinIDs,naOuts = testFlowGrd.findNonOverlappingBasins([outletRowCol])
basinMask = basinIDs == 0 #Get a mask of what is inside the first catchment

#Randomly flood this basin
randBasinMaker = rg.proceduralGrid_randomPriorityFilling(testFlowGrd,fillSlope = 1e-3,randomSeed=42,
                                                        matchMeanElevationAndRelief=True,outlets = outletRowCol,
                                                        gridMask = basinMask,
                                                         doScalePriorityByElevation=False)

#Second create an instance of the randomFilling class that assigns priorities to upstream filling randomly,
#but scaled by the current elevation
randBasinMaker_elevationScaled = rg.proceduralGrid_randomPriorityFilling(testFlowGrd,fillSlope = 1e-3,randomSeed=42,
                                                        matchMeanElevationAndRelief=True,outlets = outletRowCol,
                                                        gridMask = basinMask,
                                                         doScalePriorityByElevation=True)
#Get a random flow grid from each of these
rndFlowGrid = randBasinMaker.getFlowGridPermutation()
rndFlowGrid_elevationScaled = randBasinMaker_elevationScaled.getFlowGridPermutation()

#Create a matrix of subplots
f,axs = plt.subplots(2,3,figsize = (8,4),sharex = True, sharey = True,dpi = 120)

#In the first column plot the actual topography and drainage area
axs[0,0].set_title('Real topography',fontsize = 'small')
testFlowGrd.plotGrid(axs = axs[0,0])

axs[1,0].set_title('Real drainage area',fontsize = 'small')
testFlowGrd.plotDrainageArea(axs = axs[1,0])

#In the second column plot the random catchment created with unweighted flooding priorities
axs[0,1].set_title('Random priority flooding, elevation',fontsize = 'small')
rndFlowGrid.plotGrid(axs = axs[0,1])

axs[1,1].set_title('Random priority flooding, drainage area',fontsize = 'small')
rndFlowGrid.plotDrainageArea(axs = axs[1,1])

#In the final column, plot the random catchment created with elevation-weighted flooding priorities
axs[0,2].set_title('Random priority scaled by elevation, elevation',fontsize = 'x-small')
rndFlowGrid_elevationScaled.plotGrid(axs = axs[0,2])
axs[1,2].set_title('Random priority scaled by elevation, drainage area',fontsize = 'x-small')
rndFlowGrid_elevationScaled.plotDrainageArea(axs = axs[1,2])

plt.tight_layout()
```
