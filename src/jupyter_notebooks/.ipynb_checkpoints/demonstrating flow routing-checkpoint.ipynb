```python
'''
As a simple example, lets try and approximate how the average curvature varies on hilltops within these catchments. 

We are going to be setting the value of each pixel in the catchment to be the calculated median 'hilltop' curvature.
This is going to result in a ton of data reduncy. This is not a memory-efficient way to handle this problem, as
we really could just define polygons representing the outlines of each catchment to preform this colorcoding.

Again, this is a relatively simple example though - so we'll just use the grids we have for simplicity.

'''

#First we will create a grid to hold our 'hillslope length' information
htCurv = testDEM.duplicateGridWithNewArray(np.nan*np.zeros(testDEM.shape),newSuffix='_basinWiseHSlength')

#Lets calculate the laplacian over a relatively wide kernel to keep things smooth
crv_wide = testDEM.calcFiniteLaplacianOverWindow(N=2)

#Now we will loop through each outlet and get the median curvature on the lowest drainage area parts of the landscape
minDrainageArea = testDEM._dx*testDEM._dy #Whats the minimum drainage area for a pixel?

#Mask out those parts of the grid that have the smallest possible drainage area
lowDrainageAreaMask = flowGrid.areaGrid == minDrainageArea

htCurvArray = np.zeros(len(outlets))

for i,outlet in enumerate(outlets):
    mask = basinMask == i #Get the ith basin
    htCrvs_i = crv_wide[mask & lowDrainageAreaMask].flatten() #Mask out the slopes within this basin
    htCurvArray[i] = np.nanmedian(htCrvs_i)
    htCurv[mask] = htCurvArray[i] #Populate the area of this grid corresponding to this catchment



#Now lets actually plot these data - first plot a hillshade as a base.
ax = hs.plotGrid(cmap = 'gray', vmin = 0, vmax = 255)

#Now add the hs length plot on top
htCurv.plotGrid(axs = ax,alpha = 0.5,cmap = 'Reds_r')

#Finally, get the colorbar for the HS length data
im = ax.get_images()[-1] #Get the last image plotted
cbar = f.colorbar(im,ax = ax)
cbar.set_label(r'Hilltop $\nabla^2 z$')

```

```python
'''
Here is a slightly more complicated example of looking at variability within catchments

In the flow-length vs slope plots we can see a switch from slopes increasing with flow length to slopes decreasing
with flow length, this might represent a bulk transition from hillslopes to valleys. Lets examine how the location
of this inflection varies across these different drainage basins. 

Note, this is probably not the best way to define hillslope length. For one, this slope calculation is the slope
magnitude from a kernel that calculates slope in the x and y directions - whereas for this problem we are probably
most interested in the along-flow slope. For more information on identifying hillslope lengths see the following:

Grieve, S.W.D., Mudd, S.M., and Hurst, M.D., 2016, How long is a hillslope? 
Earth Surface Processes and Landforms, v. 41, p. 10391054, doi:10.1002/esp.3884. 

Given the grids we have, the curvature - length relationship might be a more reliable approach, but the data
has a bit more complicated structure. Since this is just a demonstration we'll use the slope-length data to keep
things simple.

'''

from scipy import optimize #Library for helping us find best-fitting parameters of a function
from scipy.interpolate import UnivariateSpline
from scipy.interpolate import CubicSpline

#First, lets define a function that will let us find the inflection point in the slope-length function above.
#It looks like (in the one example above) that we could fit two straight lines to the log-log slope area data, and
#they would joint at the peak in slopes.

def loglogsplitfit(L,theta_L,K_L,theta_R,splitLocation):
    '''
    L: lengths to preform fit at
    theta_L: exponent on power law fit, left of the split
    theta_R: exponent on power law fit, right of the split
    K_L: pre-exponential coefficient, right of the split
    splitLocation: area value at which data split
    '''
    
    #Determine which data are left and right of the split in the regression
    leftData = L <= splitLocation
    rightData = ~leftData
    
    #Preallocate space for the prediction data
    yPred = np.zeros_like(L)*np.nan
    
    #Fit the data left of the intercept
    yPred[leftData] = K_L*L[leftData]**theta_L
    
    #Find the prediction value at the split location
    y_0_r = K_L*splitLocation**theta_L 
    
    #Determine the relative location right of the split in log space (so we can treat y_0_r as the y intercept)
    logx_r_relative = np.log10(L) - np.log10(splitLocation)
    
    yPred[rightData] = 10**(np.log10(y_0_r) + theta_R*logx_r_relative[rightData])
    
    return yPred

def L2norm(L,S,params):
    '''
    Define the objective function for identifying the 'best' parameters as the sum of the squared residuals.
    
    L: maximum flow lengths
    S: slopes at those lengths
    params: list of parameters in the two segment fitted line
    '''
    theta_L,K_L,theta_R,splitLocation = params
    yPred = loglogsplitfit(L,theta_L,K_L,theta_R,splitLocation)
    
    #Divides have 0 max flow length, and thus can't be predicted in these power-law regressions
    badData = np.isnan(yPred)| np.isinf(yPred) | np.isnan(S)
    
    
    return np.sum((S[~badData]-yPred[~badData])**2)

def fitSplitFit(L,S):
    '''
    Function to actually do the minimization
    
    '''
    
    theta_L_guess = 0.1
    theta_R_guess = -0.1
    k_l_guess = 1e-1
    split_l_guess = 50.0
    paramGuess = [theta_L_guess,k_l_guess,theta_R_guess,split_l_guess] #Stick these all into a list
    
    #Define the objective function for assessing different parameter values
    objfun = lambda params: L2norm(L,S,params)
    
    return optimize.minimize(objfun,paramGuess).x #Return the best fitting parameters

'''
Lets first try this out on an example catchment from above, just to make sure its working as expected
'''

mask = basinMask == 0#Get an example catchment
s_i = slp[mask & (maxL >0)].flatten()
L_i = maxL[mask& (maxL >0)].flatten()

#Find the best fitting parameters
bfParams = fitSplitFit(L_i,s_i)

#Create an example fitted line
LtoFit = np.logspace(1,np.log10(np.max(L_i)),100) #Create a stand-in x-axis to fit
Sfit = loglogsplitfit(LtoFit,bfParams[0],bfParams[1],bfParams[2],bfParams[3])

#Lets plot this to check and see how everything looks
f,axs = plt.subplots(1,1,dpi = 120)
axs.plot(L_i,s_i,'.k',alpha =0.1,zorder = -2)
axs.plot(LtoFit,Sfit,'-',color = 'mediumseagreen', label = 'split power-law regression')

#Lets call the 'approximate hillslope length' the point where there is a break in the slope-length data
L_concConvTransition = bfParams[-1]

axs.set_yscale('log')
axs.set_xscale('log')
ylim = axs.get_ylim()

#Plot a vertical line identifying the split between slope increasing with length and slope decreasing with length
axs.plot([L_concConvTransition,L_concConvTransition], ylim,'--',color = 'orange',label = 'Aprox. hillslope length')

axs.legend(loc = 'lower left')
axs.set_xlabel('Maximum flow length')
axs.set_ylabel('Slope')
```

```python
'''
Now lets go about doing this for each catchment identified above, and color coding the catchments accordingly. 

We are going to be setting the value of each pixel in the catchment to be the calculated hillslope length (see above).
This is going to result in a ton of data reduncy. This is not a memory-efficient way to handle this problem, as
we really could just define polygons representing the outlines of each catchment to preform this colorcoding.

Again, this is a relatively simple example though - so we'll just use the grids we have for simplicity.

'''

#First we will create a grid to hold our 'hillslope length' information
hsLength = testDEM.duplicateGridWithNewArray(np.nan*np.zeros(testDEM.shape),newSuffix='_basinWiseHSlength')

#Now we will loop through each outlet and get the fitted HS length using the above procedure

#There are a lot of potential numpy warning from log transforming 0s, lets just suppress those temporarily
import warnings
warnings.filterwarnings('ignore')

#Keep a record of all the hillslope lengths measured
hsLengthArray = np.zeros(len(outlets))
medianSlopeArray = np.zeros(len(outlets)) #Lets also keep track of median slopes

for i,outlet in enumerate(outlets):
    mask = basinMask == i #Get the ith basin
    crv_i = crv_wide[mask].flatten() #Mask out the slopes within this basin
    slope_i = slp[mask].flatten()
    L_i = maxL[mask].flatten() #Mask out the flow lengths

    bfParams = fitSplitFit(L_i,slope_i) #Find the best fitting parameters for the split power law regression
    L_concConvTransition = bfParams[-1]#Last of the best fitting parameters is the split in the power law regression
    hsLength[mask] = L_concConvTransition 
    hsLengthArray[i] = L_concConvTransition
    medianSlopeArray[i] = np.nanmedian(slope_i)


#Turn warnings back on
warnings.resetwarnings()
```

```python
'''
Chi-elevation plots can be derived by first computing a chi grid.  This grid calculates Chi (e.g.,
the upstream integrated drainage area) by preforming a recursive upstream search from specified outlets. 
For more information on Chi see the following:

Perron, J.T., and Royden, L., 2013, An integral approach to bedrock river profile analysis:
Earth Surface Processes and Landforms, v. 38, p. 570576, doi:10.1002/esp.3302.


Here we will mask everything by catchment, and just look at these plots for a couple different catchments

'''

#Find all the outlets that drain to the grid boundaries
outlets = flowGrid.findBoundaryOutlets()

#Get the drainage areas for these outlets
outletAreas = flowGrid.areaGrid[outlets[:,0],outlets[:,1]]

#Lets just work on the largest two outlets for now
nBasins = 3
#Sort the drainage area, return the sorted indices, and flip to sort from large to small
srtIdcs = np.argsort(outletAreas)[::-1] 
outlets = outlets[srtIdcs[:nBasins]] #Just get the largest two outlets

#Get these basins as masks
basinMask,naOuts = flowGrid.findNonOverlappingBasins(outlets)

cmap = cm.get_cmap('tab10')
colors = [cmap(i) for i in np.linspace(0,1,nBasins)]

A_0 = 1e6 #Normalizing drainage area
theta = 0.4 # m/n exponent

#Create a couple subplots, one for the map and a few others for each basin that we'll make chi plots for
f,axs = plt.subplots(1,1+nBasins,figsize = (10,3),sharex=False,sharey = False)

#Plot a hillshade and stack the basin mask on top of that
hs.plotGrid(axs = axs[0],cmap = 'gray', vmin = 0, vmax = 255)
basinMask.plotGrid(axs = axs[0],cmap = cmap,vmin = 0, vmax = nBasins-1,alpha = 0.3)

#Loop through each of the identified catchments
for i,outlet in enumerate(outlets):
    chiGrid = flowGrid.calcChiGrid(A_0,theta,[outlet])
    
    axs[i+1].plot(chiGrid[basinMask == i].flatten()[::20],testDEM[basinMask == i].flatten()[::20],'.',
                  color = colors[i],
                 alpha = 0.05, markeredgecolor = None,markersize = 10)
    axs[i+1].set_xlabel(r'$\chi$',fontsize = 14)
    axs[i+1].set_ylabel(r'$z$',fontsize = 14)

#Make the plot a bit more compact
plt.tight_layout()

```

```python
def lengthThetaRadialDensity(lengthGrid,thetaGrid,minL = None, maxL = None, Lbins = 20,
                             dTheta = 3.1415/20, thetaWin = None,axs = None, **kwargs):
    '''
    Prototyping a function for making length-theta plots
    '''
    
    
    if thetaWin is None:
        thetaWin = dTheta/2.0
    #First construct the bins that we will do counts in
    thetaMids = np.arange(-np.pi,np.pi+dTheta,dTheta)
    
    if isinstance(Lbins,int):
        if maxL is None:
            maxL = lengthGrid.max()
        if minL is None:
            minL = np.min(lengthGrid[lengthGrid > 0])
        
        #Defaulting to log spacing here... but perhaps thats not the thing to do?
        lengthBins = np.logspace(np.log10(minL),np.log10(maxL),Lbins+1)
        
    if axs is None:
        f = plt.figure()
        axs = plt.subplot(111,projection = 'polar')
        
    #Now need to loop through the grid and count all the pixels in each direction
    
    #preallocate space for grids to count in each length theta bin, and count the total number of cells per length
    
    #Find the midpoints of each grid coordinate
    lengthMids = (lengthBins[:-1] + lengthBins[1:])/2
    
    #Preallocate space for the results grids
    binCount = np.zeros((len(lengthMids),len(thetaMids))) #A count of the number of pixels in each length-theta bin
    lengthCount = np.zeros_like(binCount) #A count of the number of pixels at each length bin, used for normalization
    
    #Loop throught the length bins - alternatively, could perhaps do this with np.histogram2d
    for i in range(len(lengthMids)):
        theseLengths = (lengthGrid >= lengthBins[i]) & (lengthGrid < lengthBins[i+1])

        theseCounts = np.zeros_like(thetaMids)
        for j in range(len(thetaMids)):
            thisThetaMid = thetaMids[j]
            
            #Find windows around this midpoint
            minTheta = thisThetaMid - thetaWin
            maxTheta = thisThetaMid + thetaWin
            
            #Need to determine whether these are appropriate
            if minTheta < -np.pi:
                minTheta = (2.0*np.pi) + minTheta
                theseDirs = (thetaGrid >= minTheta) | (thetaGrid < maxTheta)
            elif maxTheta > np.pi:
                maxTheta = -2.0*np.pi + maxTheta
                theseDirs = (thetaGrid >= minTheta) | (thetaGrid < maxTheta)
            
            else:
                theseDirs = (thetaGrid >= minTheta) & (thetaGrid < maxTheta)
            
            binCount[i,j] = np.sum(theseDirs & theseLengths)
        lengthCount[i,:] = np.sum(binCount[i,:])
            
    binCount[binCount == 0] = np.nan
    
    axs.pcolormesh(thetaMids,lengthMids,binCount/lengthCount,**kwargs)
    
    return axs, binCount, lengthCount
```
